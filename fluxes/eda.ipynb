```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
for i in range(0, 10000, 1000):
    # data = np.load(f'../examples/PyMPDATA_examples/Arabas_and_Farhat_2020/data/advectee_{i}.npy')
    data = np.load(f'./data/advectee_{i}.npy')

    plt.plot(data, alpha=1*i/10000, color='b')
plt.xlabel('x')
plt.ylabel('advectee')
plt.title('advectee evolution over time')
plt.show()
```
